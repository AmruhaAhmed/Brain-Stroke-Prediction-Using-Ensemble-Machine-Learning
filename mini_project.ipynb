```python
def clean_work_type(text):
  if text=="Never_worked":
    return "Never Worked"
  if text=="Self-employed":
    return "Self Employed"
  if text=="Govt_job":
    return "Govt Job"
  if text=="children":
    return "Children"
  if text=="Private":
    return "Private"
```

FEATURE ENGINEERING

```python
def age_group(age):
  if(age>=0 and age<=2):
    return "Baby"
  elif(age>=3 and age<=39):
    return "Young Adult"
  elif(age>=40 and age<=59):
    return "Middle Aged Adult"
  else:
    return "Old Adult"
```

```python
df['age_group']=""
```

```python
df['age_group']=df['age'].apply(age_group)
```

```python
df['age_group'].unique()
```

```python
df.columns
```

```python
def bmi_category(num):
  if num<=18.5:
    return "Underweight"
  elif num>18.5 and num<=24.9:
    return "Healthy Weight"
  elif num>25.0 and num<29.9:
    return "Overweight"
  else:
    return "Obese"
```

```python
df['bmi_category']=""
```

```python
df['bmi_category']=df['bmi'].apply(bmi_category)
```

```python
df['bmi_category'].unique()
```

```python
def diabetes_check(num):
  if num<=100.0:
    return "Normal"
  elif num>100.0 and num<125.0:
    return "Prediabetes"
  else:
    return "Diabetes"
```

```python
df['diabetes']=df['avg_glucose_level'].apply(diabetes_check)
```

```python
def diabetes_check(num):
   if num<=100.0:
    return "Normal"
   elif num>100.0 and num<125.0:
    return "Prediabetes"
   else:
    return "Diabetes"
```

```python
def bmi_category(num):
  if num<=18.5:
    return "Underweight"
  elif num>18.5 and num<=24.9:
    return "Healthy Weight"
  elif num>25.0 and num<29.9:
    return "Overweight"
  else:
    return "Obese"
```

```python
def age_group(age):
  if(age>=0 and age<=2):
    return "Baby"
  elif(age>=3 and age<=39):
    return "Young Adult"
  elif(age>=40 and age<=59):
    return "Middle Aged Adult"
  else:
    return "Old Adult"
```

```python
def age_year_convert(num):
  return (2020-num)
```
